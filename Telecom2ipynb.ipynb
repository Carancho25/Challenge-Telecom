```python
import pandas as pd
```

```python
datos = pd.read_csv("datos_limpios.csv")
```

```python
datos.head(10)
```

```python
print(datos.isnull().sum())
```

"Se elimina columna CustomerID, que no aporta para el modelo predictivo"

```python
datos = datos.drop('customerID', axis=1)
display(datos.head())
```

```python
datos.info()
```

"Identificador de columnas categoricas"

```python
categorical_cols = datos.select_dtypes(include='object').columns
print("Columnas categóricas identificadas:")
print(categorical_cols)
```

# Aplicar One-Hot Encoding a las columnas categóricas identificadas #

```python
df_encoded = pd.get_dummies(datos, columns=categorical_cols, drop_first=True)
```

# drop_first=True evita la multicolinealidad eliminando una de las columnas binarias generadas por cada variable original.#


```python
print("\nDataFrame después de One-Hot Encoding (primeras 5 filas):")
print(df_encoded.head())
print("\nColumnas del DataFrame después de One-Hot Encoding:")
print(df_encoded.columns)
print("\nDimensiones del DataFrame después de One-Hot Encoding:")
print(df_encoded.shape)
```

# Verificacion del porcentaje de cancelacion (Churn) #

## Mostrar la distribución de la variable 'Churn_numeric, para verificar si hay desbalance ##

```python
print("Distribución de la variable 'Churn_numeric':")
print(df_encoded['Churn_numeric'].value_counts())
print("\nPorcentaje de cada clase:")
print(df_encoded['Churn_numeric'].value_counts(normalize=True) * 100)
```

# II Parte:Correlacion y Seleccion de Variables #

# Analisis de Correlacion. Matriz de Correlacion #


```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = df_encoded.corr()

# Plot the heatmap
plt.figure(figsize=(18, 14))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Get correlations with 'Churn_Yes'
churn_correlations = correlation_matrix['Churn_Yes'].sort_values(key=abs, ascending=False)

# Exclude the correlation of 'Churn_Yes' with itself and 'Churn_No'
churn_correlations = churn_correlations.drop(['Churn_Yes', 'Churn_No', 'Churn_numeric'])

print("Top correlations with Churn_Yes (sorted by absolute value):")
print(churn_correlations)
```

```python
# Convert correlations to percentages and sort in ascending order
churn_correlations_percentage_asc = (churn_correlations * 100).sort_values(ascending=True)

print("Correlations with Churn_Yes (in percentages, ascending order):")
print(churn_correlations_percentage_asc)
```
